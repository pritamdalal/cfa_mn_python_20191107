# Tutorial 01 - Jupyter Notebook

The purpose of this tutorial is to introduce you to Jupyter notebook files.  This file - the one you are currently interacting with - is a Jupyter notebook.  

The notebook format conveniently allows you to combine words/sentences, computer code, code output, and mathematical notation.  Some people spend a lot of time to creating beautiful and well organized notebooks books that are used for communicating analysis (e.g. *Python Data Science Handbook* is a collection of Jupyter Notebooks).  However, a vast majority of notebooks are used simply as a convenient interative programming environment.

For those of you familiar with R, a Jupyter Notebook is similar in functionality to R markdown notebooks.  In fact, there are lots of notebook formats out there now: Matlab, Google, etc.

Even though it opens up in your browser, a Jupyter notebook does not on live on the internet, it lives on your computer.  Your web browser is simply being used as a piece of software to view and interact with the notebook.

At its heart, a notebook is a series of *cells*.  There are two kinds of cells: 1) code cells that contain (usually) python code; 2) markdown cells which contain markdown or latex.

## Edit Mode vs Command Mode

There are two modes in a notebook: 1) `edit` mode; 2) `command` mode.  

In `edit` mode you are *inside* a cell and you can edit the contents of the cell.  

In `command` mode, you are outside the cells and you can navigate between them.  

When you are in edit mode, there is a green border on the left side of the current cell; when you are in command mode the left border is blue.

## Keyboard Shortcuts

Here are some of my favorite keyboard shortcuts:

edit mode: **enter**

cmd mode: **esc**

insert cell above: **a**

insert cell below: **b**

code cell: **y**

markup cell: **m**

execute cell: **shift + enter**

navigate up: **k**

navigate down: **j**

delete cell: **d, d** (press **d** twice)
